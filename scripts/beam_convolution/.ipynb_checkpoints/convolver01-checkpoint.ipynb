```julia
function gen_Blm(res::Resolution, beammap,; weight_path)
    lmax = 3res.nside - 1
    beammap_pol = zeros(3, res.numOfPixels)
    beammap_pol[2,:] .= beammap .* (res.numOfPixels./ (4œÄ*sum(beammap)))
    Blm = hp.map2alm(beammap_pol, lmax = lmax, datapath = weight_path, use_pixel_weights=true)
    return AlmPair(Blm, conj.(Blm))
end
```

```julia
skyalm = AlmPair(alm, conj.(alm))

Blm = gen_Blm(res, beammap, weight_path="../../data/maps/healpy-data/.");
```

```julia
function convolvor(res::Resolution, sky::AlmPair, Beam::AlmPair)
    lmax = 3res.nside - 1 
    E, B = 2, 3
    """ First term: m=>0, Second term: m<0 """
    Alm·¥±Blm·¥± = sum(sky.alm[E,:] .* Beam.almconj[E,:]) + sum(sky.almconj[E,lmax+2:end] .* Beam.alm[E,lmax+2:end])
    Alm·¥ÆBlm·¥Æ = sum(sky.alm[B,:] .* Beam.almconj[B,:]) + sum(sky.almconj[B,lmax+2:end] .* Beam.alm[B,lmax+2:end])
    Alm·¥±Blm·¥Æ = sum(sky.alm[E,:] .* Beam.almconj[B,:]) + sum(sky.almconj[E,lmax+2:end] .* Beam.alm[B,lmax+2:end])
    Alm·¥ÆBlm·¥± = sum(sky.alm[B,:] .* Beam.almconj[E,:]) + sum(sky.almconj[B,lmax+2:end] .* Beam.alm[E,lmax+2:end])
    #println("Elmblmg + Blmblmc=", Elmblmg + Blmblmc)
    #println("-Elmblmc  + Blmblmg=", -Elmblmc  + Blmblmg)
    return [real(Alm·¥±Blm·¥± + Alm·¥ÆBlm·¥Æ), real(-Alm·¥±Blm·¥Æ + Alm·¥ÆBlm·¥±)]
end
```

```julia
function EffectiveBeamConvolution(res::Resolution, sky::AlmPair, beam2d, idx)
    Nx = 360
    Ny = 721
    dir = "/group/cmb/litebird/usr/ytakase/psi_db/3_30/"
    file = h5open(dir * "idx=$idx.h5", "r")
    ariaOfPixels = length(read(file,"psi"))
    convolved_map = zeros(2, res.numOfPixels)
    run_idx_range = read(file,"map_div") * 512
    
    println("ariaOfPixels : $ariaOfPixels")
    println("Run range of idx : 1 to $run_idx_range")
    p = Progress(ariaOfPixels)
    
    @time for i in 1:ariaOfPixels
        #tick()
        pixid = ariaOfPixels*(idx-1) + i
        psi_in_ipix = read(file,"psi/$i")
        effective_beam = symmetrizer(psi_in_ipix, beam2d, Nx, Ny)
        beammap = gen_beammap(res, effective_beam, pixid)       
        Blm = gen_Blm(res, beammap)
        pol_signal = convolvor(res, sky, Blm)
        convolved_map[:,pixid] = pol_signal
        next!(p)
    end
    return convolved_map
end
```
